### Handwritten digits in PCA using scikit-learning

This code performs Principal Component Analysis (PCA) on the digits dataset from scikit-learn, then visualizes the results.

Import Libraries: Imports necessary libraries like pandas for data
manipulation, matplotlib.pyplot for plotting, PCA and LDA from scikit-learn for dimensionality reduction, scale for preprocessing, cm for colormaps, and numpy for numerical operations.

Load Data: Loads the digits dataset, which is a collection of handwritten digits images.

 Data Information: Extracts the data, number of samples, number of features, number of unique digits, and the labels (the actual digit values).
Perform PCA: Initializes a PCA model with 10 components and applies it to the data to reduce its dimensionality.

Print Explained Variance: Prints the explained variance ratio for each component and the cumulative explained variance for the first two components. This tells you how much of the original data's variance is captured by each principal component.

Prepare for Plotting: Creates some sample data (X and ys) and a colormap for plotting.
Create Scatter Plot: Generates a scatter plot of the data transformed by PCA, using the first two principal components as the x and y axes. Each point is colored according to its original digit label.

Add Legend and Title: Adds a legend to the plot to identify the digits and sets a title for the plot.

Show Plot: Displays the generated plot.
In essence, the code reduces the 64-dimensional digits data to 10 dimensions using PCA and then visualizes the first two principal components to see how well the different digits are separated in this reduced space. The scatter plot helps to visually assess the effectiveness of PCA in distinguishing between the different handwritten digits.



### Code

```python
import pandas as pd
from sklearn.datasets import load_digits
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.preprocessing import scale
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA # Corrected import
import matplotlib.cm as cm
import numpy as np # Import numpy

digits = load_digits()
data = digits.data
n_samples,n_features=data.shape
n_digits=len(np.unique(digits.target))
labels = digits.target

pca = PCA(n_components=10)
data_r = pca.fit_transform(data)
print('explained variance ratio(first two components):%s'%str(pca.explained_variance_ratio_))
print('sum of explained variance(first two components):%s'%str(sum(pca.explained_variance_ratio_))) # Corrected variable name

X = np.arange(10)
ys=[i+X+(i*X)**2 for i in range(10)]
plt.figure()
colors=cm.rainbow(np.linspace(0,1, len(ys)))
# Corrected loop and variable name
for c, i, target_name in zip(colors, np.unique(labels), digits.target_names):
  plt.scatter(data_r[labels== i,0],data_r[labels== i,1],c=c, alpha =0.4, label=target_name)
plt.legend()
plt.title('Scatterplot of Points plotted in first \n''10 principle components')
plt.show() # Corrected function call
```
